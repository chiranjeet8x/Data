## Series()

```python
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7,8]
var = pd.Series(x)
var
```

#### Or

```python
x = pd.Series([4,5,6,7,8], index = ["A","B","C","D","E"])
x
```

```python
dict1 = {
    "Name" : ["Java", "C++", "C", "Python"],
    "Rank" : [9, 7, 5, 8],
    "Pop" : [12,34,44,22]
}

var = pd.Series(dict1)
var
```

```python
s = pd.Series(12)
s
```

```python
s = pd.Series(20, index = [1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
num1 = pd.Series(50, index=[1,2,3,4,5,6,7,8])
num2 = pd.Series(30, index=[1,2,3,4,5])
print(num1 + num2)
```

# Arithmetic Operation

```python
var = pd.DataFrame({
    "A" : [1,2,3,4,5],
    "B" : [8,6,4,2,1]
})

var
```

```python
var["Total A+B"] = var["A"]+var['B']
var
```

```python
var["Sub A-B"] = var["A"]-var['B']
var
```

```python
var["Div A/B"] = var["A"]/var['B']
var
```

```python
var["Mult A*B"] = var["A"]*var['B']
var
```

```python
var1 = pd.DataFrame ({
    "A" : [10,20,30,40,50,60],
    "B" : [11,22,33,44,55,66],
    "C" : [98,76,54,32,34,56],
    "D" : [22,56,100,90,300,22]
})
var1
```

```python
var1["Python"] = var1["A"] > 30
var1
```
